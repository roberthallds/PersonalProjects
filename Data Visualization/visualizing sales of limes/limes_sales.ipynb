## Visualizing Lime Sales per Species, Co-Referencing Monthly Visits to Store

A project for my Codecademy Data Science & ML Certification.

```python
from matplotlib import pyplot as plt
import warnings
import sys

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

months_adj = [0]
for i in range(len(months)):
  if i % 2 == 0:
    months_adj.append(months[i])

visits_per_month = [9695, 7909, 10831, 12942, 12495, 16794, 14161, 12762, 12777, 12439, 10309, 8724]

# numbers of limes of different species sold each month
key_limes_per_month = [92.0, 109.0, 124.0, 70.0, 101.0, 79.0, 106.0, 101.0, 103.0, 90.0, 102.0, 106.0]
persian_limes_per_month = [67.0, 51.0, 57.0, 54.0, 83.0, 90.0, 52.0, 63.0, 51.0, 44.0, 64.0, 78.0]
blood_limes_per_month = [75.0, 75.0, 76.0, 71.0, 74.0, 77.0, 69.0, 80.0, 63.0, 69.0, 73.0, 82.0]
```

```python
if not sys.warnoptions:
    import warnings
    warnings.simplefilter("ignore")

# create your figure here
plt.figure(figsize=(12, 8))
ax1 = plt.subplot(1,2,1)
x_values = range(len(months))
plt.plot(x_values, visits_per_month, color='blue', marker='*')
plt.xlabel("month")
ax1.set_xticklabels(months_adj)
plt.ylabel("visits per month")
plt.title("Monthly Store Visits for FY X")

ax2 = plt.subplot(1,2,2)
plt.plot(x_values, key_limes_per_month, color='blue', marker="o")
plt.plot(x_values, persian_limes_per_month, color='green', marker='o')
plt.plot(x_values, blood_limes_per_month, color='red', marker='o')
plt.xlabel("month")
ax2.set_xticklabels(months_adj)
plt.ylabel("limes sold per month")
plt.legend(['key limes', 'persian limes', 'blood limes'])
plt.title("Monthly Sales of Limes per Species for FY X")

plt.savefig("limesSalesAndVisits.pdf")

plt.show()
```
